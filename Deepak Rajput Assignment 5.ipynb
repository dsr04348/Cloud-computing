#### Feature Type 2 -- The doc-term matrix based on tf-idf values should satify the following demands (considering using sklearn):

5) Convert the token list per document into a text string  

##### Feature generation

2) Ignore terms that appear in more than 5 documents
3) Consider 1500 token features with top tf-idf value
4) Apply sublinear tf scaling

````python
#feature 1
doc_topics = lda.transform(docterms)

doc_topics.shape
````

````python
#feature 2
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize

#Text cleaning function
def clean_text(text):
    tokens = nltk.word_tokenize(text.lower())

    #Remove stopwords and Token length should be > 3
    stopwords = set(nltk.corpus.stopwords.words('english'))
    tokens = [token for token in tokens if token not in stopwords and len(token) > 3]

    #Porter stemming
    stemmer = nltk.stem.PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]

    #Join tokens back to get clean text
    clean_text = ' '.join(tokens)

    return clean_text
#given conditions
max_df = 0.5
min_df = 5
max_features = 1500
sublinear_tf = True

vectorizer = TfidfVectorizer(max_df=max_df, min_df=min_df, max_features=max_features, sublinear_tf=sublinear_tf, preprocessor=clean_text)

docterm_matrix = vectorizer.fit_transform(docs)

# Normalize the doc-term matrix
docterm_matrix_norm = normalize(docterm_matrix)
print(docterm_matrix_norm)
````

````python
# feature 3
from sklearn.decomposition import TruncatedSVD

lsa_model = TruncatedSVD(n_components=50, algorithm='randomized', n_iter=15, random_state=42)
lsa_features = lsa_model.fit_transform(docterm_matrix)
print(lsa_features)
````
